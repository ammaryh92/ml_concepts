```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression, Ridge, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

# Batch Gradient Descent

```python
np.random.seed(42)
```

```python
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X
```

```python
X_b = np.c_[np.ones((100,1)), X]
```

```python
theta = np.random.randn(2, 1)
```

```python
theta
```

```python
n_iterations = 1000
eta = 0.22
m = X_b.shape[0]
theta_path_bgd = []
for iteration in range(n_iterations):
    theta_path_bgd.append(theta)
    gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
```

```python
theta_path_bgd = np.array(theta_path_bgd)
```

```python
t0, t1 = np.meshgrid(np.linspace(-5, 10, 100),
                    np.linspace(-5, 10, 100))

theta_new = np.c_[t0.ravel(), t1.ravel()] 
```

```python
dict_df = {'slope':[], 'intercept':[], 'cost':[]}
for i in range(theta_new.shape[0]):
    intercept = theta_new[i, 0]
    dict_df['intercept'].append(intercept)
    slope = theta_new[i, 1]
    dict_df['slope'].append(slope)
    prediction =  intercept + slope * X
    dict_df['cost'].append((mean_squared_error(y, prediction)))
```

```python
df = pd.DataFrame(dict_df)
```

```python
z = df['cost'].values.reshape(t0.shape)
```

```python
plt.figure(figsize=(8,5))
plt.contourf(t0, t1, z, levels=[i for i in np.linspace(0, 60, 30)])
plt.axis([-2,8,-2,8])
plt.xticks(np.arange(-2,8,1))
plt.yticks(np.arange(-2,8,1))
plt.plot([4], [3], 's', color='blue', label = 'Global Minimum')
plt.plot(theta_path_bgd[0][0], theta_path_bgd[0][1], '^', color='black', label='Initial Parameters')
plt.plot(theta_path_bgd[:, 0], theta_path_bgd[:, 1], 'r-')
plt.xlabel('$Parameter 1$', fontsize = 16)
plt.ylabel('$Parameter 2$', fontsize = 16)
lgd = plt.legend(bbox_to_anchor=[1.7,1,0,0], fontsize=12)
plt.colorbar(label='Cost Function')
plt.title('Batch Gradient Descent', fontsize=16)
plt.savefig('Batch_GD.jpg', bbox_extra_artists=(lgd,), bbox_inches='tight')
```

# Stochastic Gradient Descent

### Without Learning Schedule

```python
theta = theta_path_bgd[0]
```

```python
theta_path_sgd = []
n_epochs = 50
t_0, t_1 = 5, 50
m  = X_b.shape[0]
eta = 0.22

for epoch in range(n_epochs):
    for i in range(m):
        theta_path_sgd.append(theta)
        random_idx = np.random.randint(m)
        Xi = X_b[random_idx:random_idx+1]
        yi = y[random_idx:random_idx+1]
        gradients = 2 * (Xi.T.dot(Xi.dot(theta) - yi))
        theta = theta - eta * gradients
```

```python
theta_path_sgd = np.array(theta_path_sgd)
```

```python
plt.figure(figsize=(8,5))
plt.contourf(t0, t1, z, levels=[i for i in np.linspace(0, 60, 30)])
plt.axis([-2,8,-2,8])
plt.xticks(np.arange(-2,8,1))
plt.yticks(np.arange(-2,8,1))
plt.plot([4], [3], 's', color='blue', label = 'Global Minimum')
plt.plot(theta_path_sgd[0][0], theta_path_sgd[0][1], '^', color='black', label='Initial Parameters')
plt.plot(theta_path_sgd[:, 0], theta_path_sgd[:, 1], 'r-')
plt.xlabel('$Parameter 1$', fontsize = 16)
plt.ylabel('$Parameter 2$', fontsize = 16)
lgd = plt.legend(bbox_to_anchor=[1.7,1,0,0], fontsize=12)
plt.colorbar(label='Cost Function')
plt.title('Stochastic Gradient Descent', fontsize=16)
plt.savefig('Stochastic_GD.jpg', bbox_extra_artists=(lgd,), bbox_inches='tight')
```

### With Learning Schedule

```python
t_0, t_1 = 5, 50

def learning_schedule(t):
    return t_0 / (t + t_1)
```

```python
theta = theta_path_bgd[0]
```

```python
theta_path_sgd2 = []
n_epochs = 50
m  = X_b.shape[0]

for epoch in range(n_epochs):
    for i in range(m):
        theta_path_sgd2.append(theta)
        random_idx = np.random.randint(m)
        Xi = X_b[random_idx:random_idx+1]
        yi = y[random_idx:random_idx+1]
        gradients = 2 * (Xi.T.dot(Xi.dot(theta) - yi))
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
```

```python
theta_path_sgd2 = np.array(theta_path_sgd2)
```

```python
plt.figure(figsize=(8,5))
plt.contourf(t0, t1, z, levels=[i for i in np.linspace(0, 60, 30)])
plt.axis([-2,8,-2,8])
plt.xticks(np.arange(-2,8,1))
plt.yticks(np.arange(-2,8,1))
plt.plot([4], [3], 's', color='blue', label = 'Global Minimum')
plt.plot(theta_path_sgd2[0][0], theta_path_sgd2[0][1], '^', color='black', label='Initial Parameters')
plt.plot(theta_path_sgd2[:, 0], theta_path_sgd2[:, 1], 'r-')
plt.xlabel('$Parameter 1$', fontsize = 16)
plt.ylabel('$Parameter 2$', fontsize = 16)
lgd = plt.legend(bbox_to_anchor=[1.7,1,0,0], fontsize=12)
plt.colorbar(label='Cost Function')
plt.title('Stochastic Gradient Descent with Learning Schedule', fontsize=16)
plt.savefig('Stochastic_GD_with_ls.jpg', bbox_extra_artists=(lgd,), bbox_inches='tight')
```

# Mini-Batch Gradient Descent

### Without Learning Schedule

```python
theta = theta_path_bgd[0]
theta_path_mgd = []
n_epochs = 50
minibatch_size = 20
eta = 0.25

for epoch in range(n_epochs):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        theta_path_mgd.append(theta)
        Xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = (2/minibatch_size) * (Xi.T.dot(Xi.dot(theta) - yi))
        theta = theta - eta * gradients
```

```python
theta_path_mgd = np.array(theta_path_mgd)
```

```python
plt.figure(figsize=(8,5))
plt.contourf(t0, t1, z, levels=[i for i in np.linspace(0, 60, 30)])
plt.axis([-2,8,-2,8])
plt.xticks(np.arange(-2,8,1))
plt.yticks(np.arange(-2,8,1))
plt.plot([4], [3], 's', color='blue', label = 'Global Minimum')
plt.plot(theta_path_mgd[0][0], theta_path_mgd[0][1], '^', color='black', label='Initial Parameters')
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], 'r-')
plt.xlabel('$Parameter 1$', fontsize = 16)
plt.ylabel('$Parameter 2$', fontsize = 16)
lgd = plt.legend(bbox_to_anchor=[1.7,1,0,0], fontsize=12)
plt.colorbar(label='Cost Function')
plt.title('Mini-Batch Gradient Descent', fontsize=16)
plt.savefig('Minibatch_GD.jpg', bbox_extra_artists=(lgd,), bbox_inches='tight')
```

### With Learning Schedule

```python
theta = theta_path_bgd[0]
theta_path_mgd2 = []
n_epochs = 50
minibatch_size = 20

t = 0
for epoch in range(n_epochs):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        theta_path_mgd2.append(theta)
        t+=1
        Xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = (2/minibatch_size) * (Xi.T.dot(Xi.dot(theta) - yi))
        eta = learning_schedule(t)
        theta = theta - eta * gradients
```

```python
theta_path_mgd2 = np.array(theta_path_mgd2)
```

```python
plt.figure(figsize=(8,5))
plt.contourf(t0, t1, z, levels=[i for i in np.linspace(0, 60, 30)])
plt.axis([-2,8,-2,8])
plt.xticks(np.arange(-2,8,1))
plt.yticks(np.arange(-2,8,1))
plt.plot([4], [3], 's', color='blue', label = 'Global Minimum')
plt.plot(theta_path_mgd2[0][0], theta_path_mgd2[0][1], '^', color='black', label='Initial Parameters')
plt.plot(theta_path_mgd2[:, 0], theta_path_mgd2[:, 1], 'r-')
plt.xlabel('$Parameter 1$', fontsize = 16)
plt.ylabel('$Parameter 2$', fontsize = 16)
lgd = plt.legend(bbox_to_anchor=[1.7,1,0,0], fontsize=12)
plt.colorbar(label='Cost Function')
plt.title('Mini-Batch Gradient Descent with Learning Schedule', fontsize=16)
plt.savefig('Minibatch_GD_with_ls.jpg', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
