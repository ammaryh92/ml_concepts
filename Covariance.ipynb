````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

# What is covariance?

- Covariance represents how two random variables change together.
- While variance represents the variability of one random variable, covariance represents the joint variability of two random variables.
- In other words, if two random variables (X, Y) have a positive covariance, this means that the smaller values of one variable are correlated with the smaller values of the other variable (the same applies for the greater values).

````python
X = np.arange(1,11)
````

````python
y = np.array([1,3,3,5,6,7,8,9,11,12])
````

````python
cov1 = np.cov(X, y)[0][1]
````

````python
plt.figure(figsize=(7,5))
plt.scatter(x=X, y=y)
plt.axvline(X.mean(), color='purple', linestyle='-.', label = f'Mean of X: {X.mean()}')
plt.axhline(y.mean(), color='r', linestyle='--', label = f'Mean of Y: {y.mean()}')
plt.xticks(X)
plt.yticks(np.arange(1,13));
plt.xlabel('$X$', fontsize=18)
plt.ylabel('$Y$', fontsize=18, rotation=0)
lgd = plt.legend(fontsize=12, bbox_to_anchor=(1,0.1,0.1,0.9))
props = dict(facecolor='black', shrink=1, width=0.5, headwidth=5, headlength=10)
plt.annotate('a ', (2.1,2.9), (2.5,2), arrowprops=props, fontsize=14)
plt.annotate('b', (8.1,8.9), (8.5, 8), arrowprops=props, fontsize=14)
plt.text(2.5, 9.5, 'A', fontsize=20, color='green')
plt.text(8, 9.5, 'B', fontsize=20, color='green')
plt.text(2.5, 3.5, 'C', fontsize=20, color='green')
plt.text(8, 3.5, 'D', fontsize=20, color='green')
plt.title('Covariance(X, Y) = {:0.2f}'.format(cov1), fontsize=20)
plt.savefig('images/pos_cov.jpg', bbox_extra_artists=(lgd,), bbox_inches='tight');
````

- Point a (2,3): when the X-coordinate (2) is < its mean(5.5), the Y-coordinate (3) is also < its mean (6.5).
- Point b (8,9): when the X-coordinate (8) is > its mean(5.5), the Y-coordinate (9) is also > its mean (6.5).
- So points in quadrants B and C contribute positively to the overall covariance.

- Alternatively, if two random variables (X, Y) have a negative covariance, this means that smaller values of one variable are correlated with greater values of the other variable, or greater values of one variable are correlated with smaller values of the other variable.

````python
y_reverse = np.flip(y)
````

````python
cov2 = np.cov(X, y_reverse)[0][1]
````

````python
plt.figure(figsize=(7,5))
plt.scatter(x=X, y=y_reverse)
plt.axvline(X.mean(), color='purple', linestyle='-.', label = f'Mean of X: {X.mean()}')
plt.axhline(y.mean(), color='r', linestyle='--', label = f'Mean of Y: {y_reverse.mean()}')
plt.xticks(X)
plt.yticks(np.arange(1,13));
plt.xlabel('$X$', fontsize=18)
plt.ylabel('$Y$', fontsize=18, rotation=0)
lgd = plt.legend(fontsize=12, bbox_to_anchor=(1.3,0.1,0.1,0.9))
props = dict(facecolor='black', shrink=1, width=0.5, headwidth=5, headlength=10)
plt.annotate('a', (2.1,11), (2.7,11), arrowprops=props, fontsize=14, va='center')
plt.annotate('b', (9.1,3), (9.7, 3), arrowprops=props, fontsize=14, va='center')
plt.text(2.5, 9.5, 'A', fontsize=20, color='green')
plt.text(8, 9.5, 'B', fontsize=20, color='green')
plt.text(2.5, 3.5, 'C', fontsize=20, color='green')
plt.text(8, 3.5, 'D', fontsize=20, color='green')
plt.title('Covariance(X, Y) = {:0.2f}'.format(cov2), fontsize=20)
plt.savefig('images/neg_cov.jpg', bbox_extra_artists=(lgd,), bbox_inches='tight');
````

- Point a (2,11): when the X-coordinate (2) is < its mean(5.5), the Y-coordinate (11) is also > its mean (6.5).
- Point b (9,3): when the X-coordinate (9) is > its mean(5.5), the Y-coordinate (3) is also < its mean (6.5).
- So points in quadrants A and D contribute negatively to the overall covariance.

In general, covariance can classify 3 types of relationships:

    a. relationships with a positive trend (positive covariance)
    b. relationships with a negative trend (negative covariance)
    c. No relationship (covariance is near zero).

````python
y_2 = np.array([5,12,4,9,1,10,6,9,3,9])
````

````python
cov3 = np.cov(X, y_2)[0][1]
````

````python
plt.figure(figsize=(7,5))
plt.scatter(x=X, y=y_2)
plt.axvline(X.mean(), color='purple', linestyle='-.', label = f'Mean of X: {X.mean()}')
plt.axhline(y_2.mean(), color='r', linestyle='--', label = f'Mean of Y: {y_2.mean()}')
plt.xticks(X)
plt.yticks(np.arange(1,13));
plt.xlabel('$X$', fontsize=18)
plt.ylabel('$Y$', fontsize=18, rotation=0)
lgd = plt.legend(fontsize=12, bbox_to_anchor=(1.3,0.1,0.1,0.9))
plt.text(2.5, 9.5, 'A', fontsize=20, color='green')
plt.text(8, 9.5, 'B', fontsize=20, color='green')
plt.text(2.5, 3.5, 'C', fontsize=20, color='green')
plt.text(8, 3.5, 'D', fontsize=20, color='green')
plt.title('Covariance(X, Y) = {:0.2f}'.format(cov3), fontsize=20)
plt.savefig('images/no_cov.jpg', bbox_extra_artists=(lgd,), bbox_inches='tight');
````

- The covariance is very close to zero which means there is no visible linear correlation between X and Y variables.
- Points which contribute positively to the overall covariance in quadrants B and C are cancelled out by points which contribute negatively to the covariance in quadrants A and D.

# Disadvantages of Covariance

- Covariance value is difficult to interpret.
- Although covariance is useful for examining the existence of linear correlation between two random variables, it is difficult to use covariance to quantify the strength of the relationship.
- While a higher covariance value (positive or negative) usually indicate a stronger linear correlation, that is not always the case.

````python
plt.figure(figsize=(12,5))
plt.subplot(121)
y_3 = 2 * X + 3
cov4 = np.cov(X, y_3)[0][1]
sns.regplot(x=X, y=y_3)
plt.xlabel('$X$', fontsize=18)
plt.ylabel('$Y$', fontsize=18, rotation=0, labelpad=10)
plt.title('Case 1: Covariance(X, Y) = {:0.2f}'.format(cov4), fontsize=15);
plt.subplot(122)
df = pd.read_csv('datasets/cov1.csv')
df = df.sample(20)
X_2, y_4 = df['x'], df['y']
cov5 = np.cov(X_2, y_4)[0][1]
sns.regplot(x=X_2, y=y_4)
plt.xlabel('$X$', fontsize=18)
plt.title('Case 2: Covariance(X, Y) = {:0.2f}'.format(cov5), fontsize=15)
plt.savefig('images/cov_cases.jpg');
````

- Notice that while X, and Y variables have a stronger linear correlation in case 1, their covariance value in case 2 is much higher.
- This is because the covariance value is highly affected by the scale of the data.
- As you increase the scale of the data as in case 2, the covariance value increases dramatically.

The value of the covariance can take any value between (-, ) depending on:

    a. Whether the slope is positive or negative.
    b. The variance of each variable.
    c. The scale of the data.

# Solution

- One solution to the data scaling problem is to divide the covariance value by the standard deviation of each variable.
- This effectively eliminates the difference in data scales squeezes the covariance down to a value between -1 and 1.
- The new value is called the correlation coefficient.
