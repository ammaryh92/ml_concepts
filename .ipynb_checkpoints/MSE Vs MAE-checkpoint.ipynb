```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
X = np.random.randn(400)
```

```python
y = 4 * X + 2
```

```python
slopes = np.linspace(-5, 15, 500)
```

```python
squared_errors = []
absolute_errors = []
for s in slopes:
    preds = s * X + 2
    squared_errors.append(mean_squared_error(y, preds))
    absolute_errors.append(mean_absolute_error(y, preds))
squared_errors = np.array(squared_errors)
absolute_errors = np.array(absolute_errors)
```

```python
plt.figure(figsize=(8,6))
plt.plot(slopes, squared_errors, label = 'Mean Squared Error')
plt.plot(slopes, absolute_errors, label = 'Mean Absolute Error')
plt.xticks(np.arange(-5, 15, 1))
plt.legend(fontsize=14)
plt.ylabel('Cost Function', fontsize=14)
plt.xlabel('Parameter', fontsize = 14)
plt.title('MSE Vs MAE', fontsize = 18)
plt.savefig('MSE_Vs_MAE.jpg')
```
